# 03 - Preparación de datos

En este notebook se realiza la limpieza y transformación del dataset con el objetivo de dejar los datos listos para el modelado.3

```python
import pandas as pd

df = pd.read_csv("../data/raw/student_dataset.csv")
```

## Transformación de la variable objetivo

Se convierte la variable objetivo en una variable binaria para facilitar el modelado.

- 1 → Dropout
- 0 → No Dropout

```python
df["Target_binary"] = df["Target"].apply(lambda x: 1 if x == "Dropout" else 0)
```

## Eliminación de la variable original

Se elimina la columna original Target para evitar redundancia.

```python
df = df.drop("Target", axis=1)
```

## Separación de variables

Se separan las variables independientes (X) y la variable objetivo (y).

```python
X = df.drop("Target_binary", axis=1)
y = df["Target_binary"]
```

## Generación del dataset limpio

Se guarda el dataset procesado para ser utilizado en la etapa de modelado.

```python
df.to_csv("../data/processed/clean_dataset.csv", index=False)
```
