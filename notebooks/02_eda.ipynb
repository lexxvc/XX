# 02 - Análisis exploratorio de datos (EDA)

En este notebook se realiza un análisis exploratorio de los datos con el objetivo de identificar patrones, distribuciones y posibles relaciones entre variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/student_dataset.csv")
```

## Distribución de la variable objetivo

Se analiza la proporción de estudiantes en cada categoría.

```python
df["Target"].value_counts().plot(kind="bar")
plt.title("Distribución del Target")
plt.show()
```

## Distribución de la edad

Se analiza la distribución de la edad de los estudiantes.

```python
sns.histplot(df["Age at enrollment"], kde=True)
plt.title("Distribución de la edad")
plt.show()
```

## Género vs abandono

Se analiza la relación entre el género y la deserción.

```python
sns.countplot(x="Gender", hue="Target", data=df)
plt.title("Género vs Target")
plt.show()
```

## Correlación entre variables

Se analiza la relación entre variables numéricas mediante una matriz de correlación.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Matriz de correlación")
plt.show()
```
